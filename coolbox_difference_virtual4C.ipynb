```python
import coolbox
from coolbox.api import *
import cooler
import bioframe
import tabix 
import numpy as np
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib.patches as patches
from matplotlib.ticker import EngFormatter
```

```python
#plot a matrice with ChIP and annotation tracks
with MinValue(-7), MaxValue(-3):
    cool1 = Cool('PATH_TO/A.cool', style='triangular')

chip_bigwig = f"track.bw"

regions= ["Chr:X-Y"]

frame = XAxis() + \
    cool1 + Title("title") + Color('viridis_r') + \
    InsuScore(cool1, max_value=0.7, min_value=-0.7, color='darkslateblue') + Title("Insulation score") + \
    Spacer(0.5) + \
    BigWig(chip_bigwig) + Color("mediumpurple") + Title("title") + MinValue(0) + MaxValue(40000) + \
    Spacer(0.5) + \
    GTF(f"genome_annotation.gtf", name_attr="gene_id", length_ratio_thresh=0.0001) + \
    Spacer(0.5)
frame *= Feature(depth_ratio=0.35)
fig = frame.plot("Chr:X-Y")
fig.savefig("output_file_name.svg")
frame.plot("Chr:X-Y")
```

```python
#difference plot

cool1 = Cool('PATH_TO/A.cool', style='triangular')
cool2 = Cool('PATH_TO/B.cool', style='triangular')

bigwig = f"PATH_TO/A.bw"
dnt1_bw=f"PATH_TO/B.bw"

regions= ["Chr:X-Y"]

frame = XAxis() + \
    HiCDiff(cool1, cool2, normalize='none', cmap="coolwarm_r", style="triangular") + \
    MinValue(-1.5) + MaxValue(1.5)  + \
    Spacer(0.5) + \
    BigWig(bigwig) + Color("teal") + Title("title") + MinValue(0) + MaxValue(260000) + \
   Spacer(0.5) + \
    GTF(f"PATH_TO/genome_annotation.gtf", name_attr="gene_id", length_ratio_thresh=0.0001) + \
    BigWig(dnt1_bw) + Color("purple") + Title("Dnt1") + MinValue(0) + MaxValue(1700000) + \
    Spacer(2)
frame *= Feature(depth_ratio=0.35)
fig = frame.plot("Chr:X-Y")
fig.savefig("output_file_name.svg")
frame.plot("Chr:X-Y")
```

```python
#vitual 4C

cool1 = Cool('PATH_TO/A.cool')
cool2 = Cool('PATH_TO/B.cool')


dnt1_bw = f"PATH_TO/C.bw"

regions= ["Chr:X-Y"]

frame = XAxis() + \
    Spacer(0.5) + \
    Virtual4C(cool1, "Chr:K-L",  min_value=-0.0005, max_value=0.01, style='filled') + \
    Spacer(0.5) + \
    Virtual4C(cool1, "Chr:K-L", style='heatmap') + \
    Spacer(0.5) + \
    Virtual4C(cool2, "Chr:K-L", min_value=-0.0005, max_value=0.01, style='filled') + \
    Spacer(0.5) + \
    Virtual4C(cool2, "Chr:K-L", style='heatmap') + \
    Spacer(0.5) + \
    GTF(f"PATH_TO/genome_annotation.gtf", name_attr="gene_id", length_ratio_thresh=0.0001) + \
    BigWig(dnt1_bw) + Color("purple") + Title("Dnt1") + MinValue(0) + MaxValue(1700000) + \
    Spacer(0.5)
frame *= Feature(depth_ratio=0.7)
frame = frame * Vlines(regions, line_width=3)
fig = frame.plot("I:1-2000")
fig.savefig("output_file_name.svg")
frame.plot("Chr:X-Y")
```
